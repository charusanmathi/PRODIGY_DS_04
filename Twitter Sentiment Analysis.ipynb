```python
# CONVERTING ALL THE TWEET INTO LOWER STRING DATA

# Regex is used to remove all special characters as it hinders digitation
```

```python
import re
```

```python
#Text transformation into Lowercase and removing all special characters

# FOR THE TRAIN DATASET

# Lower case transforamtion
df_train["lower"]=df_train.tweet.str.lower() 

# Converting all to string
df_train["lower"]=[str(data) for data in df_train.lower] 

# Regex function to remove special characters
df_train["lower"]=df_train.lower.apply(lambda x: re.sub('[^A-Za-z0-9 ]+', ' ', x)) 

# FOR THE VALIDATION DATASET

# Lower case transforamtion
df_val["lower"]=df_val.tweet.str.lower() 

# Converting all to string
df_val["lower"]=[str(data) for data in df_val.lower] 

# Regex function to remove special characters
df_val["lower"]=df_val.lower.apply(lambda x: re.sub('[^A-Za-z0-9 ]+', ' ', x)) 
```

```python
# DATA WRANGLING

# LABEL ENCODING


from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()

df_train['type'] = le.fit_transform(df_train['type'])
```

```python
df_train
```

```python
X = df_train['lower']
```

```python
y = df_train['type']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train,y_test = train_test_split(X,y, test_size = 0.3,  random_state  = 43, stratify=df_train['type'])
```

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
clf = Pipeline([
    ('vectorizer_tri_grams', TfidfVectorizer()),
    ('random_forest', (RandomForestClassifier()))         
])
```

```python

clf.fit(X_train,y_train)
```

```python
clf_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import *
```

```python
print(accuracy_score(y_test, clf_pred))
```

```python

# Print the classfication report
print(classification_report(y_test, clf_pred))
```
